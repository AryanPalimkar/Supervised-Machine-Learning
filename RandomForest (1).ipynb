```python
import pandas as pd
import numpy as np
db=pd.read_csv("/content/heart_cleveland_upload.csv")
db.head(5)
#thalach stands for maximum heart rate
```

```python
print("The shape of the dataset is:",db.shape)
```

```python
db.describe()
```

```python
label=np.array(db['condition'])#getting the target values
params=db.drop('condition', axis=1)#creating a new dataframe after dropping the column axis
params_list=list(params.columns)#extracting the list of features
params=np.array(params)
```

```python
from sklearn.model_selection import train_test_split
train_params,test_params, train_labels, test_labels=train_test_split(params, label, test_size=0.25, random_state=12)
print("The shape of the training dataset is :",train_params.shape)
print("The shape of the testing dataset is :", test_params.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
#creating a random forest classifier object 
clf=RandomForestClassifier(n_estimators=100)
#fitting refers to training the model using training data
clf.fit(train_params, train_labels)
#predicting the output of the model for the test data
pred_labels=clf.predict(test_params)

```

```python
from sklearn import metrics
print("The accuracy of the model is:", metrics.accuracy_score(pred_labels,test_labels)
)
```

```python
print("The confusion matrix for the model is:",metrics.confusion_matrix(pred_labels,test_labels))
```

```python
#Increasing the number of tree ensembles might help in accuracy
clf2=RandomForestClassifier(n_estimators=128)
#fitting refers to training the model using training data
clf2.fit(train_params, train_labels)
#predicting the output of the model for the test data
pred_labels=clf2.predict(test_params)
print("The accuracy of the new model is:",metrics.accuracy_score(pred_labels,test_labels)
)
```
